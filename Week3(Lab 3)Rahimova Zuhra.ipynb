```python
Factor = 2
scaling_matrix = cv2.getRotationMatrix2D((0, 0), 0,Factor)
scaled_image = cv2.warpAffine(image,
scaling_matrix, (w*Factor, h*Factor))
```

```python
plt.imshow(cv2.cvtColor(scaled_image,
cv2.COLOR_BGR2RGB))
plt.title('Scaled Image (2x)')
plt.axis('off')
plt.show()
print(scaled_image.shape)
```

```python
factor = 5
M_task3 = cv2.getRotationMatrix2D((0, 0), 0, factor)
scaled_task3 = cv2.warpAffine(image, M_task3, (w * factor, h * factor))

cv2.imwrite('scaled_5x.png', scaled_task3)

plt.imshow(cv2.cvtColor(scaled_task3, cv2.COLOR_BGR2RGB))
plt.title('Task 3: Scaled 5x (Check Pixelation)')
plt.axis('off')
plt.show()
```
